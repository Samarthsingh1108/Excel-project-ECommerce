```python
!pip install googletrans==4.0.0-rc1 ipywidgets
import ipywidgets as widgets
from googletrans import Translator, LANGUAGES
from IPython.display import display

translator = Translator()


input_label = widgets.Label(value="Enter text:")
input_text = widgets.Textarea(placeholder="Type the text to translate here...", layout=widgets.Layout(width='80%'))
output_label = widgets.Label(value="Translated text:")
output_text = widgets.Textarea(value="", layout=widgets.Layout(width='80%'), disabled=True)


language_options = list(LANGUAGES.values())
language_dropdown = widgets.Dropdown(
    options=language_options,
    value='english',
    description='Choose language:',
)


translate_button = widgets.Button(
    description="Translate",
    button_style='success',
)


display(input_label, input_text)
display(language_dropdown)
display(translate_button)
display(output_label, output_text)


def on_translate_button_click(b):
    try:
        selected_lang = language_dropdown.value
        lang_code = list(LANGUAGES.keys())[list(LANGUAGES.values()).index(selected_lang)]
        translated = translator.translate(input_text.value, dest=lang_code)
        output_text.value = translated.text
    except Exception as e:
        output_text.value = f"Error: {e}"

translate_button.on_click(on_translate_button_click)
```

```python
from gtts import gTTS
from IPython.display import Audio, display, FileLink
import ipywidgets as widgets


text_input = widgets.Textarea(placeholder="Enter text to convert to speech...", layout=widgets.Layout(width='80%', height='100px'))
convert_button = widgets.Button(description="Convert to Speech", button_style='success')

display(text_input, convert_button)


def on_button_click(b):
    text = text_input.value
    if text.strip():

        tts = gTTS(text=text, lang='en')
        output_path = "/content/output.mp3"
        tts.save(output_path)


        display(Audio(output_path, autoplay=True))


        display(FileLink(output_path))
    else:
        print("Please enter some text!")

convert_button.on_click(on_button_click)
```
